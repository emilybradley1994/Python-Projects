```python
from IPython.display import clear_output
import random

def display_board(board):

    print ("  " + board[1] + "  |  " + board[2] + "  |  " + board[3] + "  ")
    print ("-----------------")
    print ("  " + board[4] + "  |  " + board[5] + "  |  " + board[6] + "  ")
    print ("-----------------")
    print ("  " + board[7] + "  |  " + board[8] + "  |  " + board[9] + "  ")
    
    pass


def player_input():
    marker = ""
    while marker != "X" and marker != "O":
        marker = input("Player 1, would you like to be X or O?\n\n")
    
    if marker == "X":
        return ('X', 'O')
    
    else:
        return ('O', "X")
    
    
def place_marker(board, marker, position):
    
    board[position] = marker
    
    pass


def win_check(board, mark):
    
    return (board[1] == board[2] == board[3] == mark) or (board[4]==board[5]==board[6]==mark) or (board[7]==board[8]==board[9]==mark) or (board[2]==board[5]==board[8]==mark) or (board[3]==board[6]==board[9]==mark) or (board[1]==board[4]==board[7]==mark) or (board[1]==board[5]==board[9]==mark) or (board[3]==board[5]==board[7]==mark)
    
    pass


def choose_first():
    if random.randint(0,1) == 0:
        return "Player 1"
    else:
        return "Player 2"
    
    pass


def space_check(board, position):
    
    return board[position] == ""



def full_board_check(board):
    for x in range(1,10):
        if space_check(board, x) == 'X' or space_check(board, x) == 'O':
            return True
        else:
            return False
    
    pass


def player_choice(board):
    position = 0
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board, position):
        try:
            position = int(input ("Please choose your next position between 1-9:\n\n"))
        except:
            print ("Oops - that was not an integer between 1-9!\n\n")
        else:
            if not space_check(board, position):
                print ("Oops - this position is not available.\n\n")
    
    return position


def replay():
    return input("Would you like to play again?\n\n")[0].lower() == 'y'
    
    pass




    
while True:
    
    print('Welcome to Tic Tac Toe!\n\n')
    TheBoard = ["","","","","","","","","",""]
    display_board(TheBoard)
    (player1, player2) = player_input()
    clear_output()
    print ("Player 1, you will be {}, Player 2 will be {}\n\n".format(player1, player2))
    first_player = choose_first()
    print("{} will begin!\n\n".format(first_player))
    if input ("Are you ready to begin the game?\n\n")[0].lower() == "y":
        game_on = True
        clear_output()
        display_board(TheBoard)
    else:
        break
        
    while game_on == True:

        if first_player == 'Player 1':
            position = player_choice(TheBoard)
            clear_output()
            place_marker(TheBoard, player1, position)
            
            if win_check(TheBoard, player1):
                display_board(TheBoard)
                print ("\n\nCongratulations - Player 1 has won the game!\n\n")
                break
            elif full_board_check(TheBoard):
                display_board(TheBoard)
                print ("It's a draw!\n\n")
                break
            else:
                display_board(TheBoard)
                first_player = "Player 2"
            

        else:
            position = player_choice(TheBoard)
            clear_output()
            place_marker(TheBoard, player2, position)
            
            if win_check(TheBoard, player2):
                display_board(TheBoard)
                print ("\n\nCongratulations - Player 2 has won the game!\n\n")
                break
            elif full_board_check(TheBoard):
                display_board(TheBoard)
                print ("It's a draw!\n\n")
                break
            else:
                display_board(TheBoard)
                first_player = "Player 1"

    
    if not replay():
        break

        
        
print ("Thanks for playing!")
            
        
        

```
